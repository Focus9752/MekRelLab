```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def E_b(A, Z):
    # Neutrontallet
    N = A - Z
    
    # MeV
    a1 = 15.86
    a2 = 18.34
    a3 = 0.714
    a4 = 23.21
    a5 = 0

    if N % 2 == 0 and Z % 2 == 0:
        a5 = 12.00
    elif A % 2 != 0:
        a5 = 0
    else:
        a5 = -12.00

    return a1 * A - a2 * A**(2/3) - a3 * ((Z*(Z-1))/(A**(1/3))) - a4*(((N-Z)**2)/A) + a5 * A**(-1/2)
```

Beregn bindingsenergien af 1H, 56Fe og 235U.

```python
print(E_b(1, 1))

print(E_b(56, 26))

print(E_b(235, 92))
```

Gennemsnitlig bindingsenergi af 45Fe-72Fe

```python
A_arr = []
E_b_arr = []

for i in range(45, 72 + 1):
    A_arr.append(i)
    E_b_arr.append(E_b(i, 26))
    
plt.figure(1)
plt.grid()
plt.title("Nukleontal og bindingsenerg")
plt.plot(A_arr, E_b_arr)
```
